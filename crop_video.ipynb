```python
import os
import shutil
import zipfile

from moviepy import VideoFileClip
from tqdm.notebook import tqdm
```

```python
full_clip = VideoFileClip("./lovely runner ep 1.mp4")
```

```python
import json
sample = json.load(open("./sample.json", "r"))
```

```python
sample
```

```python
duration = full_clip.duration
```

```python
full_clip.cropped()
```

```python
for scene in sample:
    clip = full_clip.subclipped(max(0, int(scene["start_time"])), min(int(scene["end_time"]), duration))
    clip.write_videofile(f"./sample/lovely_runner_scene_{scene['scene_id']}.mp4", fps=30, threads=1, codec="libx264")
```
